```python
# Exercise 1: Make Plurals
# Write a program to generate plural forms of English nouns from a given list of singular nouns.
# Simplification rules:
# 1. Add 's' to most words (e.g., "cat" -> "cats").
# 2. If the word ends with 'y' and is preceded by a consonant, replace 'y' with 'ies' (e.g., "baby" -> "babies").
# 3. If the word ends with 'f' or 'fe', replace it with 'ves' (e.g., "wolf" -> "wolves").
# 4. Add 'es' to words ending in 's', 'x', 'z', 'ch', or 'sh' (e.g., "box" -> "boxes").
# 5. For irregular nouns, handle exceptions (e.g., "child" -> "children", "man" -> "men").

nouns = ["cat", "baby", "wolf", "box", "child", "man", "mouse", "knife", "bush"]
plurals = []

# Dictionary for irregular nouns (Rule 5)
irregular_nouns = {
    "child": "children",
    "man": "men",
    "mouse": "mice"
}

for word in nouns:
    # Rule 5: Check for irregular nouns first
    if word in irregular_nouns:
        plurals.append(irregular_nouns[word])
    
    # Rule 3: Words ending in 'f' or 'fe'
    elif word.endswith(('f', 'fe')):
        if word.endswith('f'):
            plurals.append(word[:-1] + "ves")
        else:
            plurals.append(word[:-2] + "ves")
            
    # Rule 2: Ends with 'y' preceded by a consonant
    elif word.endswith('y') and word[-2] not in "aeiou":
        plurals.append(word[:-1] + "ies")
        
    # Rule 4: Ends with 's', 'x', 'z', 'ch', or 'sh'
    elif word.endswith(('s', 'x', 'z', 'ch', 'sh')):
        plurals.append(word + "es")
        
    # Rule 1: Default case, just add 's'
    else:
        plurals.append(word + "s")

print(plurals)
```
